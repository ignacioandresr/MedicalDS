Testeo de API /record (Historial Médico)

```python
endpoint_records = f"{BASE_URL}/records"
print(f"Intentando acceder a: {endpoint_records}")

try:
    response = requests.get(endpoint_records)
    response.raise_for_status()

    data = response.json()
    status_code = response.status_code

    print(f"\n✅ Petición Exitosa (Código: {status_code})")
    print(f"Tipo de datos recibidos: {type(data)}")

    if isinstance(data, list):
        print(f"Total de elementos: {len(data)}")
        display(pd.DataFrame(data).head(3))
    else:
        print("Respuesta (JSON):")
        display(data)

except requests.exceptions.RequestException as e:
    print(f"\n❌ Error en la Petición a /records: {e}")
    if 'response' in locals() and response.content:
        print(f"Contenido de la Respuesta de Error:\n{response.text}")
```
